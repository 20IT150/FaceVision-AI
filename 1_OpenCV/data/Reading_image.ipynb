```python
import numpy as np
import  cv2
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.image as matimg
```

```python
img = cv2.imread('./flying_bird.jpg')
img2  = matimg.imread('./flying_bird.jpg')
img3 = Image.open('./flying_bird.jpg')
```

```python
print(img)
```

```python
plt.imshow(img)
```

```python
plt.figure(figsize=(100,30))

plt.subplot(1,3,1)
plt.imshow(img)
plt.axis('off')
plt.title("BGR")


plt.subplot(1,3,2)
plt.imshow(img2)
plt.axis('off')
plt.title("RBG1")


plt.subplot(1,3,3)
plt.imshow(img3)
plt.axis('off')
plt.title("RGB2")

plt.show()

```
